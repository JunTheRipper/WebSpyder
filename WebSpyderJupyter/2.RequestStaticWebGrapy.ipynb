下面介绍一下使用content抓取二进制图片等等并写入文件的流程

**具体：爬取简书的图标**

```python
headers ={'User-Agent':'Chrome/10'}
r = requests.get('https://www.jianshu.com/favicon.ico',headers=headers)
try:
    r.raise_for_status()
    r.encoding = r.apparent_encoding
    print("显示文本：",r.text)
    with open("1.ico",'wb') as file:
        file.write(r.content)
        print("成功写入!")
except Exception as e:
    print(e)
```

很明显，输入二进制最后显示的r.text数据是一团乱码，

但是r.content可以显示二进制数据，同时可以写入文件中.

**同理，音频视频文件也是以这种方式读取！**

requests 还提供了证书验证的功能。当发送 HTTP 请求的时候，它会检查SSL证书，
我们可以使用verify参数控制是否检查此证书。其实如果不加verify参数的话，默认是True，会自动验证。

证书没有被官方 CA 机构信任，会出现证书验证错误的结果。

```python
#4.1 测试直接爬取谷歌   出现报错SSLERROR
r = requests.get('https://www.google.com')
try:
    r.raise_for_status() #抛出异常问题
    r.encoding = r.apparent_encoding
    print(r.status_code)
    print(r.text)
except Exception as e:
    print(e)
```

```python
#尝试解决方案：把 verify 参数设置为 False 即可
headers = {'User-Agent':'Chrome/10'}
r = requests.get('https://www.google.com',verify=False,headers=headers)
try:
    r.raise_for_status() #抛出异常问题
    r.encoding = r.apparent_encoding
    print(r.status_code)
    print(r.text)
except Exception as e:
    print(e)
```

此外还有一种使用Request()对象方式执行操作,下面模拟执行POST请求

```python
from requests import Session,Request
url = 'http://httpbin.org/get'
data = {'name':'germey'}
headers={'User-Agent':'Chrome/10'}
s = Session()
req = Request('POST',url,headers,data=data)
p = s.prepare_request(req)
r = s.send(p)
print(r.text)
```

正则表达式模块（简述）
==

正则表达式中仅仅了解compile()方法  注意贪婪和非贪婪匹配问题

**compile()方法可以将正则字符串编译成正则表达式对象，实现匹配复用**

```python
import re
s = 'wise@123().s?{}Python'
pattern = r'\w{5}'
p = re.findall(pattern,s,re.I)
print(p)

print(re.match(pattern,s,re.I))
stn = re.search(pattern,s,re.M)
print(stn)  #返回对象属性
print(stn.group()) #匹配的值有哪些？
print("匹配字符串长度",stn.span())
pat = r'\w+'
print(re.search(pat,s,re.I))
print(re.findall(pat,s,re.I))
```

```python
import requests
from selenium import webdriver
from pyquery import PyQuery as pq
key ="如何掌控你的自由时间"
baiduUrl = "https://www.baidu.com/s?wd=%s site:open.163.com"
headers = {'User-Agent':'Chrome/10'}
def autoget():
    browser = webdriver.Edge()
    browser.get(baiduUrl % key)
    print(browser.page_source)


def staticget():
    res = requests.get(baiduUrl,headers=headers)
    try:
        res.raise_for_status()
        res.encoding = res.apparent_encoding
        page = pq(res.text)
        for i in page.items():
            print(i)
    except Exception as e:
        print(e)

if __name__ =='__main__':
    staticget()
```
